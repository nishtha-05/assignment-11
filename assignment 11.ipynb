```python
1.Custom exceptions provide you the flexibility to add attributes and 
methods that are not part of a standard exception class.
These can store additional information, like an application-specific error code,
or provide utility methods that can be used to handle or present the exception to a user.

example->
class validate(Exception):
    def __init__(self ,msg):
        self.msg = msg
    
def validateage(age):
    if age<0:
        raise validate("negative age")
    elif age<100:
        raise validate("perfect age")
    else:
        raise validate("age is very high")
try:
    age= int(input("enter your age"))
    validateage(age)
except validate as e:
    print(e)
        
```
